#เปรียบเทียบความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของนักประเมินแต่ละคนและของราคาตั้งขาย

ใช้ MAPE เป็นเกณฑ์ในการเปรียบเทียบ

hint เริ่มต้นมาเราอยากได้ตารางรูปแบบนี้ก่อน

image.png

```python
from sklearn.metrics import mean_absolute_percentage_error as mape #mape
```

##บ้าน

```python
est2['ประเมินโดย 1'] = est2['post id'].map(est1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
est2.head()
```

```python
est2.rename(columns = {'ประเมินราคาซื้อ-ขาย' : 'ประเมินโดย 2'}, inplace = True) #เปลี่ยนชื่อ
```

```python
all = est2[['post id', 'ราคาทรัพย์', 'ประเมินโดย 1', 'ประเมินโดย 2']]
all
```

```python
all.isnull().any()
```

```python
all = all.dropna(subset=['ประเมินโดย 2'])
all
```

```python
all.isnull().any()
```

##MAPE ของบ้าน

```python
actual_h = all['ราคาทรัพย์']
pred1_h = all['ประเมินโดย 1']
pred2_h = all['ประเมินโดย 2']
```

```python
mh1 = mape(actual_h, pred1_h)
print(f'ราคาทรัพย์เทียบกับผู้ประเมินคนที่ 1 มีค่า MAPE เท่ากับ {mh1}')
```

```python
mh3 = mape(pred1_h, pred2_h)
print(f'ผู้ประเมินคนที่ 1 เทียบกับผู้ประเมินคนที่ 2 มีค่า MAPE เท่ากับ {mh3}')
```
